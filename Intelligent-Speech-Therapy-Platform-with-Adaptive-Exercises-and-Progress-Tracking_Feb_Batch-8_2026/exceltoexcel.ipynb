```python
import pandas as pd
from g2p_en import G2p

g2p = G2p()

# Load Excel
df = pd.read_excel("D:\INFOSYS\Sentence.xlsx")

def sentence_to_phonemes(sentence):
    phonemes = g2p(sentence)
    phonemes = [p for p in phonemes if p != ' ']
    return " ".join(phonemes)

# Convert
df["phonemes"] = df["Sentence"].apply(sentence_to_phonemes)

# Save output
df.to_excel("Sentence_with_phonemes.xlsx", index=False)
print(" Phoneme conversion complete")
```
