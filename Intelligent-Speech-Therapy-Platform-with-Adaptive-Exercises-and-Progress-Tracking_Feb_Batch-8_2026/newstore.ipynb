```python
import pandas as pd
from g2p_en import G2p

# Load Excel
df = pd.read_excel("D:\SPEECH\Sentence_with_phonemes.xlsx")
df = df.dropna(subset=["Sentence", "phonemes"])
g2p = G2p()

word_level_data = []

for index, row in df.iterrows():

    sentence = row["Sentence"]
    words = sentence.split()

    for word in words:

        phonemes = g2p(word)

        # remove spaces and punctuation
        phonemes = [p for p in phonemes if p != " "]

        phoneme_string = " ".join(phonemes)

        word_level_data.append({
            "sentence": sentence,
            "word": word,
            "phonemes": phoneme_string
        })

# Create new dataframe
word_df = pd.DataFrame(word_level_data)

# Save new Excel
word_df.to_excel("D:\SPEECH\word_level_dataset.xlsx", index=False)

print(" Word-level dataset created successfully!")
```
